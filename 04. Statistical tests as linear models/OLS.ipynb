```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
 
np.random.seed(42069)
 
# Пусть у нас есть набор данных, где есть линейная зависимость y от X1 и X2
# При этом X1 и X2 тоже малясь зависимые
df = pd.DataFrame({'x1': np.random.uniform(0, 10, size=2000)})
df['is_control'] = np.array([0]*1000+[1]*1000)
df['const'] = 1
 
```

```python
df.groupby('is_control').x1.mean()
```

```python
from scipy.stats import ttest_ind

ttest_ind(df[df.is_control == 0].x1, df[df.is_control == 1].x1, equal_var = False)
```

```python
# Построим линейную регрессию-МНК из 1, X1 и X2
model = sm.OLS(
    endog=df['x1'],
    exog=df[['const', 'is_control']]
).fit()

# Внимательно смотрим на коэффициент при X2
model.summary()
```

```python
model.pvalues['is_control']
```

```python
from scipy.stats import mannwhitneyu
```

```python
mannwhitneyu(df[df.is_control == 0].x1, df[df.is_control == 1].x1)
```

```python
from scipy.stats import rankdata
```

```python
df['x1_rank'] = rankdata(df.x1)
```

```python
# Построим линейную регрессию-МНК из 1, X1 и X2
model = sm.OLS(
    endog=df['x1_rank'],
    exog=df[['const', 'is_control']]
).fit()

# Внимательно смотрим на коэффициент при X2
model.pvalues['is_control']
```
